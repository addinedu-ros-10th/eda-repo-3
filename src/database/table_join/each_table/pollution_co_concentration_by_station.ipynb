```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
import seaborn as sns
from sqlalchemy import create_engine
from dotenv import load_dotenv
import os


# ▒ .env 환경변수 로드
load_dotenv()
host = os.getenv("DB_HOST")
user = os.getenv("DB_USER")
password = os.getenv("DB_PASSWORD")
database = os.getenv("DB_NAME")

# ▒ DB 연결
engine = create_engine(f"mysql+pymysql://{user}:{password}@{host}:3306/{database}")
```

```python
# 데이터 조회: 서울특별시의 자치구별 월별 CO 농도
query = """
SELECT 
    station_name AS 자치구,
    january, february, march, april, may, june, 
    july, august, september, october, november, december
FROM pollution_co_concentration_by_station
WHERE province = '서울'
"""

df = pd.read_sql(query, engine)
df
```

```python



# 월별 long format으로 변환
df_melted = df.melt(id_vars='자치구', 
                    value_vars=['january', 'february', 'march', 'april', 'may', 'june',
                                'july', 'august', 'september', 'october', 'november', 'december'],
                    var_name='월', value_name='CO농도(ppm)')

# 월 영문 → 한글 매핑
month_mapping = {
    'january': '1월', 'february': '2월', 'march': '3월', 'april': '4월',
    'may': '5월', 'june': '6월', 'july': '7월', 'august': '8월',
    'september': '9월', 'october': '10월', 'november': '11월', 'december': '12월'
}
df_melted['월'] = df_melted['월'].map(month_mapping)

# 자치구 이름 기준 정렬
df_melted = df_melted.sort_values(by='자치구')

df_melted
```

```python
# 월별 long format 데이터(300행)를 자치구별 집계 데이터(25행)로 변환

# 기존 월별 데이터 (df_melted)
# df_melted: 자치구, 월, CO농도(ppm) 컬럼을 가진 300행 데이터프레임

# 자치구별 CO농도 집계 (평균 계산)
df_district_summary = df_melted.groupby('자치구')['CO농도(ppm)'].mean().reset_index()
df_district_summary.columns = ['자치구', 'CO농도_평균(ppm)']

# 평균값 반올림 (소수점 3자리)
df_district_summary['CO농도_평균(ppm)'] = df_district_summary['CO농도_평균(ppm)'].round(3)

# 자치구 이름 기준 정렬
df_district_summary = df_district_summary.sort_values(by='자치구')

# 인덱스 초기화
df_district_summary = df_district_summary.reset_index(drop=True)

print(f"변환 완료: {len(df_district_summary)}개 자치구")
print(df_district_summary)
```
